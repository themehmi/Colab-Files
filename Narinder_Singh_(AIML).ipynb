###Functions

In Python, a function is a reusable block of code that performs a specific task, defined using the def keyword and executed only when called.

```python
#print prime numbers using user defined function
def is_prime(start_range, end_range):
  for num in range(start_range,end_range+1):
    if num > 1:
      is_prime = True
      for i in range(2, int(num**0.5) + 1):
        if (num % i) == 0:
          is_prime = False
          break
      if is_prime:
        print(num)
        is_prime = True
start_range = 1
end_range = 10
is_prime(start_range, end_range)
```

```python
#check if a number is an armstrong by defining user defined function
def is_armstrong(num):
  power=len(__builtins__.str(num))
  total=0
  while num>0:
    digit=num%10
    total=total**power
    num//=10
  if total==num:
   is_armstrong=True
  else:
   is_armstrong=False
  return is_armstrong
print(is_armstrong(153))
```

```python
#check if a str is a palindrome
def is_palindrome(str):
  if str==str[::-1]:
    is_palindrome=True
  else:
    is_palindrome=False
  return is_palindrome
print(is_palindrome("dead"))
```

```python
# 28.Write a Python Program to check if given array is Monotonic.
#29.A monotonic array is one that is entirely non-increasing or non-decreasing.
def is_monotonic(arr):
  increasing=decreasing=True
  for i in range(1, len(arr)):
        if arr[i] > arr[i-1]:
            decreasing = False
        if arr[i] < arr[i-1]:
            increasing = False
            return increasing or decreasing
arr1 = [1, 2, 2, 3]
arr2 = [6, 5, 4, 4]
arr3 = [1, 3, 2]

print(arr1,":",is_monotonic(arr1))
print(arr2,":",is_monotonic(arr2))
print(arr3,":",is_monotonic(arr3))
```

```python
#52.Write a Python program to find all duplicate characters in string
def find_duplicates(s):
    seen = __builtins__.set()
    duplicates = __builtins__.set()
    for char in s:
        if char not in seen:
            seen.add(char)
        else:
            duplicates.add(char)
    return __builtins__.list(duplicates)
print(find_duplicates("Hello"))
```

**Converting Text Labels to Numbers**

Machine learning models work with numbers, not text. We convert flower names into numeric labels.



```python
url = 'https://raw.githubusercontent.com/themehmi/datasets/refs/heads/main/Iris1.csv'
# convert in numpy array
import numpy as np
raw_data = np.genfromtxt(url, delimiter=",", dtype=str)
# Remove empty rows
raw_data = raw_data[~np.any(raw_data == '', axis=1)]
X = raw_data
print(X)
```

```python
print("Dataset shape:", X.shape)
print("First 5 rows:\n", X[:5])
```

Converting Text Labels to Numbers

Machine learning models work with numbers, not text. We convert flower names into numeric labels.

```python
X = raw_data[:, :4].astype(float)
y_text = raw_data[:, 4]
print(y_text)
classes, y = np.unique(y_text, return_inverse=True)
print("Classes:", classes)
print("Numeric labels (first 10):", y[:10])
```

Boolean Masking (Filtering Data by Class)

```python
setosa = X[y == 0]
versicolor = X[y == 1]
virginica = X[y == 2]

print("Setosa samples:", setosa.shape)
print("Versicolor samples:", versicolor.shape)
print("Virginica samples:", virginica.shape)
```

Feature Normalization
Normalization makes features comparable in scale.

```python
X_mean = X.mean(axis=0)
X_std = X.std(axis=0) + 1e-8

X_norm = (X - X_mean) / X_std

print("Mean after normalization:", X_norm.mean(axis=0))
print("Std after normalization:", X_norm.std(axis=0))
```

Correlation Matrix
Correlation tells us how features are related.

```python
corr = np.corrcoef(X_norm.T)
print(corr)
```
